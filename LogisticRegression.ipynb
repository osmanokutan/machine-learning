# Logistic Regression

In (multinomial) logistic regression, class prediction involves the use of an affine map. If there are $p$ features and $m$ classes, an affine map in the form of $x \mapsto Ax+b$, where $A$ is a matrix of shape $(m, p)$ and $b$ is a $p$-dimensional vector, is employed for class prediction through the expression:
$$x \mapsto \mathrm{argmax}_{j=1,\dots,p}(Ax+b)_j$$
The affine map can also be leveraged to determine class probabilities as follows: We define the probability of an object with features $x$ belonging to class $j$ as
$$\frac{e^{(Ax+b)_j}}{\sum_{k=1}^me^{(Ax+b)_k}} $$
To simplify these expressions, we introduce the matrix $W$ of shape $(m, p+1)$, obtained by appending $b$ to $A$ as its last column, and denote $\tilde{x}$ as $x$ with an additional $1$ as its last entry. With these notations, the above probability can be rewritten as:
$$\frac{e^{(W\tilde{x})_j}}{\sum_{k=1}^me^{(W\tilde{x})_k}}$$

## Loss function

To learn an effective affine map from a given dataset $(X, y)$, we define a loss function $\mathcal{L}(W)$ and seek to minimize this loss by finding suitable values for $W$. For each $(x_i, y_i)$ in the dataset, we aim for the probability assigned to $x_i$ belonging to class $y_i$ to be close to one. Therefore, the negative logarithm of that probability serves as an appropriate loss function. Consequently, our loss function is expressed as:
$$\mathcal{L}(W) = \frac{1}{n}\sum_{i=1}^n -\log\bigg(\frac{e^{(W\tilde{x}_i)_{y_i}}}{\sum_{k=1}^me^{(W\tilde{x}_i)_k}}\bigg)  $$

To analyze the gradient of this function, we introduce the following variables:

- $u_i=(u_{i,1},\dots,u_{i,m})$, where $u_{i,j}=(W\tilde{x}_i)_j$
- $z_i=(z_{i,1},\dots,z_{i,m})$, where $z_{i,j}= e^{u_{i,j}}/\sum_{k=1}^m e^{u_{i,k}}$

The loss function can be expressed as:
$$\mathcal{L}(W) = \frac{1}{n}\sum_{i=1}^n-\log(z_{i,y_i}).$$

We can derive the following partial derivatives:
\begin{equation*}
    \begin{split}
        \frac{\partial u_{i,j}}{\partial W_{k,l}}&=\delta^{j,k}\tilde{x}_{i,l} \\
        \frac{\partial z_{i,j}}{\partial u_{i,k}}&=z_{i,j}(\delta^{j,k}-z_{i,k})
    \end{split}
\end{equation*}
where $\delta^{j,k}=1$ if $j=k$ else $0$. Combining these through chain rule, we obtain

\begin{equation*}
    \begin{split}
        \frac{\partial z_{i,j}}{\partial W_{k,l}} &= z_{i,j}(\delta^{j,k}-z_{i,k})\tilde{x}_{i,l}
    \end{split}
\end{equation*}

Finally, we have
\begin{equation*}
    \begin{split}
        \frac{\partial \mathcal{L}}{\partial W_{k,l}} &= -\frac{1}{n} 
        \sum_{i=1}^n (\delta^{y_i,k}-z_{i,k})\tilde{x}_{i,l}
    \end{split}
\end{equation*}

To express this in matrix form, we introduce the following notation:

- $\tilde{X}$ represents the augmented matrix of shape $(n,p+1)$ obtained from $X$ by adding an additional column with $1$'s.
- $P$ denotes the matrix of shape $(n,m)$ where $P_{i,j}=1$ if $y_i=j$ and $0$ otherwise.
- $Z$ denotes the matrix of shape $(n,m)$ where $Z_{i,j}=e^{(W\tilde{x}_i)_j}/\sum_{k=1}^me^{(W\tilde{x}_i)_k} $

Then, we can express the gradient of the loss function as:
$$\frac{\partial \mathcal{L}}{\partial W} = \frac{-1}{n}(P-Z)^T \tilde{X}. $$

# Implementation

```python
def augment(X):
    return np.concatenate((X,np.ones((X.shape[0],1))),axis=1)
def gradient(X_aug,P,W):
    Z = softmax(X_aug@W.T)
    return -((P-Z).T @ X_aug)/len(y), loss(P,Z)
def loss(P,Z):
    return -(P*np.log(Z)).sum()/P.shape[0]
def oneHotEncode(y):
    return 1.0*np.array([np.arange(y.max()+1)==x for x in y])
def softmax(U):
    E=np.exp(U).T
    return (E/E.sum(axis=0)).T
```

```python
class LogisticRegressor:
    def __init__(self):
        self.weights = None
        self.loss = None
    def fit(self,X,y,lr=0.1,num_iter=1000):
        self.weights = np.zeros((y.max()+1,X.shape[1]+1))
        self.loss = np.zeros(num_iter)
        X_aug = augment(X)
        P = oneHotEncode(y)
        for i in range(num_iter):
            grad, l = gradient(X_aug,P,self.weights)
            self.loss[i]=l
            self.weights -= lr*grad
    def predict(self,X):
        return (augment(X)@self.weights.T).argmax(axis=1)
```

# Example

```python
#create and visualize the data set
figure, (ax1, ax2) = plt.subplots(figsize=(10,5),nrows=1,ncols=2)
np.random.seed(13)
N=500
X = np.random.uniform(0,1,(N,2))
y = 1*(X.sum(axis=1)>0.7)+1*(X@np.array([0.5,1.5])>1.2)
sns.scatterplot(x=X[:,0],y=X[:,1],hue=y,ax=ax1)
ax1.set_title('Classification Data Set')

#fit and visualize a logistic regressor
lr = LogisticRegressor()
lr.fit(X,y,lr=1,num_iter=10000)
y_pred = lr.predict(X)
acc = (y == y_pred).mean()
sns.scatterplot(x=X[:,0],y=X[:,1],hue=y_pred,ax=ax2)
ax2.set_title(f'Logistic regressor predictions, training accuracy {acc:0.2f}');
```
