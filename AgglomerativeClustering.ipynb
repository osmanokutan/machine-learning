```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
from scipy.cluster.hierarchy import dendrogram
from heapq import heapify, heappop, heappush
```

# Agglomerative Clustering

In agglomerative clustering, one clusters a given data set $(x_i)_{i=1}^N$ iteratively as follows:
- Initialize $N$ clusters $\{x_i\}$ for $i=1,\dots,N$
- Iteratively, merge the two clusters which are the most similar to each other.

Note that at the end of the $K_{th}$ iteration, we have $N-K$ clusters. For measuring the similarity between two clusters, here we use the following metrics:
- Single linkage: $$d_{S}(A,B):= \min_{a \in A, b \in B} ||a-b|| $$
- Complete linkage: $$d_{C}(A,B):= \max_{a \in A, b \in B} ||a-b||$$
- Average linkage: $$d_{A}(A,B):= \frac{1}{|A||B|} \sum_{a \in A, b \in B} ||a-b||$$ Here $|A|,|B|$ denotes the number of elements in $A,B$ respectively.
- Ward linkage: $$d_{W}(A,B):= \sum_{x \in A \cup B}||x-\mu_{A \cup B}||^2 - \sum_{x \in A} ||x-\mu_A||^2 - \sum_{x \in B} ||x-\mu_B||^2$$
Here $\mu_A,\mu_B,\mu_{A \cup B}$ denotes the means of $A, B, A \cup B$ respectively.

## Recursive computation of linkage distances

If we use the definition of the linkage distance to compute it, it becomes more costly as the cluster sizes increase. Here, we show that we can calculate linkage distance recursively, taking advantage of the distances calculated before. Below we derive recursive formulas for computing singe linkage, complete linkage and average linkage distances. We are assuming $A_1,A_2,B$ are disjoint.

$$\begin{equation*}
    \begin{split}
        d_S(A_1 \cup A_2,B)&=\min_{a \in A_1 \cup A_2, b \in B} ||a-b|| \\
         &= \min \bigg( \min_{a \in A_1, b \in B} ||a-b||, \min_{a \in A_2, b \in B} ||a-b|| \bigg) \\
         &= \min \bigg(d_S(A_1,B), d_S(A_2,B) \bigg) \\
    \end{split}
\end{equation*}$$

$$\begin{equation*}
    \begin{split}
        d_C(A_1 \cup A_2,B)&=\max_{a \in A_1 \cup A_2, b \in B} ||a-b|| \\
         &= \max \bigg( \max_{a \in A_1, b \in B} ||a-b||, \max_{a \in A_2, b \in B} ||a-b|| \bigg) \\
         &= \max \bigg(d_C(A_1,B), d_C(A_2,B) \bigg) \\
         &= \max_{i=1,2, \, j=1,\dots,m} d_C(A_i,B_j) \\
         &\geq d_C(A_1,A_2).
    \end{split}
\end{equation*}$$

$$\begin{equation*}
    \begin{split}
        d_A(A_1 \cup A_2,B)&=\frac{1}{(|A_1|+|A_2|)|B|}\sum_{a \in A_1 \cup A_2, b \in B} ||a-b|| \\
         &= \frac{1}{(|A_1|+|A_2|)|B|}\left( \sum_{a \in A_1, b \in B} ||a-b|| + \sum_{a \in A_2, b \in B} ||a-b|| \right) \\
         &= \frac{1}{(|A_1|+|A_2|)|B|}\bigg(|A_1||B|d_A(A_1,B)+|A_2||B|d_A(A_2,B) \bigg) \\
         &= \frac{|A_1|}{|A_1|+|A_2|}d_A(A_1,B) + \frac{|A_2|}{|A_1|+|A_2|}d_A(A_2,B)
    \end{split}
\end{equation*}$$

Finally, let us obtain a recursive formula for Ward linkage.
To analyze the Ward distance, given a set $C$ in $\mathbb{R}^n$, we define $S(C)$ by
$$S(C):= \sum_{x \in C} ||x-\mu_C||^2 $$
With this notation, we have
$$d_W(A,B)=S(A \cup B) - S(A) - S(B). $$
Assume $C$ is the disjoint union of $C_1,\dots,C_n$. Let $\mu$ denote the center of $C$, and $\mu_i$ denote the center of $C_i$. We have
$$\begin{equation*}
    \begin{split}
        S(C)&=\sum_{i=1}^n \sum_{x \in C_i}||x-\mu||^2 \\
            &=\sum_{i=1}^n \sum_{x \in C_i} ||x-\mu_i||^2+||\mu_i-\mu||^2 + 2(x-\mu_i)\cdot(\mu_i-\mu) \\
            &=\sum_{i=1}^n S(C_i) + \sum_{i=1}^n |C_i| \,||\mu_i-\mu||^2 + \sum_{i=1}^n \left(\sum_{x \in C_i} (x-\mu_i) \right) \cdot (\mu_i-\mu) \\
            &= \sum_{i=1}^n S(C_i) + |C| \sum_{i=1}^n \frac{|C_i|}{|C|}||\mu_i-\mu||^2 \\
            &= \sum_{i=1}^n S(C_i) + |C| \sum_{i,j: i<j} \frac{|C_i||C_j|}{|C|^2} ||\mu_i - \mu_j||^2 \\
            &= \sum_{i=1}^n S(C_i) + \sum_{i,j: i<j} \frac{|C_i||C_j|}{|C|} || \mu_i - \mu_j||^2
    \end{split}
\end{equation*}$$
Using this, we get
$$d_W(A,B) = \frac{|A||B|}{|A|+|B|}|| \mu_A - \mu_B||^2 $$
$$S(C) = \sum_{i} S(C_i) + \sum_{i,j: i<j} \frac{|C_i|+|C_j|}{|C|}d_W(C_i,C_j) $$
Now, we can derive the following formula:
$$\begin{equation*}
    \begin{split}
        d_W(A_1\cup A_2,B)&=S(A_1 \cup A_2 \cup B) - S(A_1 \cup A_2) - S(B) \\
                &= S(A_1 \cup A_2 \cup B) - S(A_1)- S(A_2) - S(B) - \big(S(A_1\cup A_2)-S(A_1)-S(A_2) \big) \\
                &= \frac{|A_1|+|B|}{|A_1|+|A_2|+|B|}d_W(A_1,B) + \frac{|A_2|+|B|}{|A_1|+|A_2|+|B|}d_W(A_2,B) - \frac{|B|}{|A_1|+|A_2|+|B|}d_W(A_1,A_2)
    \end{split}
\end{equation*}$$

## Monotonicity and dendrogram representation of clustering

Agglomerative clustering can be represented by a binary tree whose nodes are the clusters formed during the algorithm as follows. The initial singleton clusters are the leaves, and each cluster formed later are considered to be the parent of the clusters forming them.

A linkage distance is called **monotone**, if the linkage distances calculated during iterations form a non-decreasing sequence. In this case, we can plot the binary three, where leaves are plotted at height $0$, and each node is plotted at the height corresponding to the associated linkage distance. Such a graphical representation is called the **dendrogram** of the clustering.

To show that all of the linkage distances we introduced are monotone, we first establish the following sufficient condition:

**Theorem** Let $d$ be a linkage distance such that for each disjoint $A_1,A_2,B$ satisfying $d(A_1,A_2) \leq d(A_1,B), d(A_2,B)$, one has $d(A_1 \cup A_2,B) \geq d(A_1,A_2)$. Then $d$ is monotone.

**Proof**
Let us show that clusters formed at later iterations have higher linkage distance associated to them. It is enough to show that if $A_1,A_2$ are merged at $n_{th}$ iteration, and $B,C$ are merged at $(n+1)_{th}$ iteration, then $d(A_1,A_2) \leq d(B,C)$. If both $B$ and $C$ are distinct from $A_1 \cup A_2$, then they were present during the $n_{th}$ iteration, hence $d(A_1,A_2) \leq d(B,C)$. Now, we can assume that $C=A_1 \cup A_2$. Note that $B$ was present during the $n_{th}$ iteration, so $d(A_1,A_2) \leq d(A_1,B), d(A_2,B)$. By the assumption in the statement of the theorem, we have $d(B,C)=d(A_1 \cup A_2,B) \geq d(A_1,A_2)$. This completes the proof.

By the recursive formulas we obtained for single, complete, average and Ward linkage, one can see that they all satisfy the condition of the theorem above, hence they are monotone.

# Implementing Agglomerative Clustering

```python
class AgglomerativeClustering:
    def __init__(self, method='ward'):
        self.method=method
    def fit(self,X):
        #initiate the distances between clusters
        N=X.shape[0]
        self.num_points=N
        M=np.zeros((2*N-1,2*N-1))
        for i in range(N):
            for j in range(i):
                d=np.linalg.norm(X[i]-X[j])
                M[i,j]=d
                M[j,i]=d
        if self.method == 'ward':
            M=M**2/2
        #use a heap to get the closest clusters during iterations
        distance_heap = [(M[i,j],i,j) for j in range(1,N) for i in range(j)]
        heapify(distance_heap)
        merged=[False]*(2*N-1)
        linkage = np.zeros((N-1,4)) #to keep track of clusters and corresponding linkage distances
        for i in range(N-1):
            #find closest clusters and add them to merged
            while True:
                dist, c1, c2 = heappop(distance_heap)
                if not (merged[c1] or merged[c2]):
                    break
            merged[c1]=True
            merged[c2]=True
            #find the number of elements of the new cluster
            num1 = linkage[c1-N,3] if c1>=N else 1
            num2 = linkage[c2-N,3] if c2>= N else 1
            num = num1 + num2
            #update linkage
            linkage[i,0]=c1
            linkage[i,1]=c2
            linkage[i,2]=dist
            linkage[i,3]=num
            #update distances
            r1=num1/num
            r2=num2/num
            for c in range(i+N):
                if not merged[c]:
                    if self.method == 'average':
                        d=r1*M[c1,c]+r2*M[c2,c]
                    elif self.method == 'single':
                        d=min(M[c1,c],M[c2,c])
                    elif self.method == 'complete':
                        d=max(M[c1,c],M[c2,c])
                    elif self.method == 'ward':
                        num3 = linkage[c-N,3] if c>=N else 1
                        total = num1+num2+num3
                        d=((num1+num3)/total*M[c1,c]
                           +(num2+num3)/total*M[c2,c]
                           -num3/total*M[c1,c2])
                    M[c,i+N]=d
                    M[i+N,c]=d
                    heappush(distance_heap,(d,c,i+N))
            self.linkage=linkage
    def predict(self,num_clusters=1,level=float('inf')):
        #cluster the data set used for training, depending on the number of clusters or linkage distance level,
        #depending on which is achieved first during the clustering
        N=self.num_points
        clusters={i:[i] for i in range(N)}
        for i in range(N-1):
            if self.linkage[i,2]>level or i>=N-num_clusters:
                break
            #get clusters to merge
            c1, c2 = self.linkage[i,:2]
            #form the new cluster
            new_cluster=clusters[c1]+clusters[c2]
            #remove merged clusters
            clusters.pop(c1)
            clusters.pop(c2)
            #add new clusters
            clusters[i+N]=new_cluster
        y=np.zeros(N).astype(int)
        for i, c in enumerate(clusters):
            for j in clusters[c]:
                y[j]=i
        return y
```

# Example

```python
#Create and visualize a data set
M=50
r=0.17
np.random.seed(13)
X=np.concatenate((
    r*np.random.randn(M,2)+np.array([0,0]),
    r*np.random.randn(M,2)+np.array([1,1]),
    r*np.random.randn(M,2)+np.array([0,1]),
),axis=0)

sns.scatterplot(x=X[:,0],y=X[:,1]);
```

```python
#Fit clustering models
model_single = AgglomerativeClustering(method='single')
model_complete = AgglomerativeClustering(method='complete')
model_average = AgglomerativeClustering(method='average')
model_ward = AgglomerativeClustering(method='ward')

model_single.fit(X)
model_complete.fit(X)
model_average.fit(X)
model_ward.fit(X)
```

### Dendrogram of single linkage clustering

```python
#Plot the dendrogram
plt.figure(figsize=(10,5))
dendrogram(model_single.linkage);
plt.title('Dedrogram of the single linkage clustering');
```

### Dendrogram and clusters of complete linkage clustering

```python
#Plot the dendrogram
plt.figure(figsize=(10,5))
dendrogram(model_complete.linkage);
plt.title('Dedrogram of the complete linkage clustering')
plt.xticks([]);
```

### Dendrogram and clusters of average linkage clustering

```python
#Plot the dendrogram
plt.figure(figsize=(10,5))
dendrogram(model_average.linkage);
plt.title('Dedrogram of the average linkage clustering')
plt.xticks([]);
```

### Dendrogram and clusters of Ward linkage clustering

```python
#Plot the dendrogram
plt.figure(figsize=(10,5))
dendrogram(model_ward.linkage);
plt.title('Dedrogram of the Ward linkage clustering')
plt.xticks([]);
```

## Choosing the number of clusters

One can take advantage of the dendrogram to determine the number of clusters. As we move up along the dendrogram, change in the number of clusters slows down and at one point becomes stable for a long time. When this happens, the corresponding number of clusters gives a good hint about a good choice of number of clusters. Also, the strength of this stability can give us confidence about the clustering method. In our examples, the stability is best observed with average linkage and Ward linkage, which suggests 3 clusters.

## Clusters

```python
fig, axs = plt.subplots(2,2,figsize=(10,10))

#Get clusters
y=model_single.predict(num_clusters=3)
sns.scatterplot(x=X[:,0],y=X[:,1],hue=y,ax=axs[0,0]);
axs[0,0].set_title('3 clusters obtained by single linkage clustering')

#Get 3 clusters
y=model_complete.predict(num_clusters=3)
sns.scatterplot(x=X[:,0],y=X[:,1],hue=y, ax=axs[0,1]);
axs[0,1].set_title('3 clusters obtained by complete linkage clustering')

#Get 3 clusters
y=model_average.predict(num_clusters=3)
sns.scatterplot(x=X[:,0],y=X[:,1],hue=y, ax=axs[1,0]);
axs[1,0].set_title('3 clusters obtained by average linkage clustering')

#Get 3 clusters
y=model_ward.predict(num_clusters=3)
sns.scatterplot(x=X[:,0],y=X[:,1],hue=y, ax=axs[1,1]);
axs[1,1].set_title('3 clusters obtained by Ward linkage clustering');
```

# Observations

We observe that single linkage exhibits poor performance. While complete linkage performs better, it produces an unnatural cluster. Both single linkage and complete linkage fail to consider the statistical properties of the point distribution. Consequently, even a small number of points where clusters become very near or very far from each other can result in unnatural clusters. This is not observed in average and Ward linkage methods, as they aggregate contributions from individual points. As a result, we find that average and Ward linkage methods perform significantly better.
